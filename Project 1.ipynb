```python
import pandas as pd
```

```python
import numpy as np
```

```python
#Assigning employee details from csv to the employee_details variable
employee_details=pd.read_csv("Employee Details - Sheet1.csv")
employee_details
```

```python
#Table related to DA is read from CSV file
DA = pd.read_csv("DA.csv")
DA
```

```python
#Joining the two tables via join method from pandas
result = employee_details.join(DA)
result
```

```python
#Converting it into dataframes
df = pd.DataFrame(result) 
df
```

```python
#Dropping some of the irrevelant columns
df = df.drop(['Join Date','Designation Code'],axis=1)
df
```

```python
df = df.drop(['Designstion Code'],axis=1)
df
```

```python
#Filling the values of unwanted data i.e, NaN to 0
df = df.fillna(0)
df
```

```python
#Renaming the columns for better access
df.columns = ['Emp_No','Emp_Name','Department','basic','hra','IT','Designation','DA']
df
```

```python
#Salary count as per the given question
salary_count = df['basic']  - df ['IT'] + df['DA'] +df['hra']
df['salary'] = salary_count
df
```

```python
#Dropping the irrevelant data and final table is shown
df=df.drop(['basic','hra','IT','DA'],axis=1)
df
```

```python
#Greater than 1001 Employee Number
result = df.loc[df['Emp_No']>1001]
result
```

```python
#When we need only 1001 detail
result = df.loc[df['Emp_No']==1001]
result
```
